```python
check_id = netflix_df['show_id'].duplicated().value_counts()
check_id
```

```python
#customized index
netflix_df.set_index('show_id', inplace= True)
netflix_df.head(2)
```

```python
#seperate seasons and time from duration column
netflix_df["duration"]=netflix_df["duration"].astype(str)
```

```python
netflix_df["seasons"] = netflix_df["duration"]
```

```python
netflix_df["duration"].mask(netflix_df["duration"].str.contains("Seasons"),"",inplace=True)
netflix_df["duration"].mask(netflix_df["duration"].str.contains("Season"),"",inplace=True)
```

```python
netflix_df["seasons"].mask(netflix_df["seasons"].str.contains("min"),"",inplace=True)
```

```python
#customized column
netflix_df= netflix_df[["type","title","director","country","cast","date_added","release_year","seasons","duration","rating","listed_in"]]
netflix_df.head(2)
```

```python
#check how many null value present in each column
netflix_df.isnull().sum()
```

```python
#showing null value by using heatmap
sns.heatmap(netflix_df.isnull())
```

```python
netflix_df.fillna(" ", inplace= True )
netflix_df
```
