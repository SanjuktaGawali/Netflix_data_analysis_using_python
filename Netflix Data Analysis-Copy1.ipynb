### Find all the instances where:
#Type(category) is movie and type.1 is dramas
or
#type(category) is TV Shows and type.1 is Kids TV

```python
#Show movie that falls under the drama type
#By using & operator
netflix_df[(netflix_df['type']=='Movie') & (netflix_df['type.1']=='Dramas')]
```

```python
#Show TV show that falls under the Kids' TV
#By using & operator
netflix_df[(netflix_df['type']=='TV Show') & (netflix_df['type.1']=="Kids' TV")]
```

```python
#combined previous two conditions using or operator
netflix_df[(netflix_df['type']=='Movie') & (netflix_df['type.1']=='Dramas')| (netflix_df['type']=='TV Show') & (netflix_df['type.1']=="Kids' TV")]
```
